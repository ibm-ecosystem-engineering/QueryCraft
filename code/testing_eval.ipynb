```python
import db2_connector
import ibm_db
```

```python
conn = db2_connector.db2_connector()
```

```python
stmt = ibm_db.exec_immediate(conn, "SELECT * FROM people")

# Fetch results
result = ibm_db.fetch_assoc(stmt)

# Process fetched rows
while result:
    # Do something with the fetched row
    print(result)
    
    # Fetch the next row
    result = ibm_db.fetch_assoc(stmt)



# Close the connection
ibm_db.close(conn)
```
